Let me provide you with an example of how to apply a simple grayscale filter on an image.

When transforming an RGB image into grayscale, we calculate the average of the RGB values for each pixel. The array holding the grayscale image has two dimensions, unlike the RGB image which has three. This distinction arises because in a grayscale image, the three RGB values are substituted with a single average value derived from the Red, Green, and Blue values of the pixel.

```python
grayscale_image = []
# Iterating Over Rows of Pixels in the Image
for i in range(len(img_array)):
    temp = []
    # Getting The Values from Each Pixel from The Row
    for j in range(len(img_array[i])):
        # Get the Average Value of RGB Values from the Pixel and append it to a temporary list
        avg = int(sum(img_array[i][j])/3)
        temp.append(avg)
    # append the row of average pixel values to a new array
    grayscale_image.append(temp)
```
